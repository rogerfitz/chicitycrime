```python
# Read in our data. We've aggregated it by date already, so we don't need to worry about paging
query = ("https://data.cityofchicago.org/resource/ijzp-q8t2.json?"
         "$where=primary_type='HOMICIDE'"
         "&$select=date,block,arrest,community_area,district,ward,latitude,longitude,location_description"
         )
print query
crimes = pd.read_json(query)
```
