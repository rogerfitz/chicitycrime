```python
files = []
for y in range(2005,2015):
    for m in range(1,10):
        f = 'chicitycrime/raw_data/{0}0{1}.csv'.format(y,m)
        files.append(f)
    for m in range(10,13):
        f = 'chicitycrime/raw_data/{0}{1}.csv'.format(y,m)
        files.append(f)
```

```python
months = []
for f in files:
    month = pd.read_csv(f, skiprows=7, error_bad_lines=False)
    month = month[month.Tmin.notnull()]
    month = month[['YearMonthDay','Tmax', 'Tmin', 'Tavg', 'Depart','PrecipTotal','Heat']]
    months.append(month)
```

```python
weather = pd.concat(months, axis=0, join='outer', ignore_index=True)
weather
```

```python
weather.to_csv(path_or_buf='chicitycrime/data/weather.csv')
```
